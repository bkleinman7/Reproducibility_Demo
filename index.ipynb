```r
library(dplyr)
library(ggplot2)
library(tidyverse)
library(readxl)
library(httr)
url1 = "https://datadryad.org/bitstream/handle/10255/dryad.108959/Data%20for%20figure%202.xlsx?sequence=1"
GET(url1, write_disk(tf <- tempfile(fileext = ".xlsx")))

data_a = read_excel(tf, 1)
#data_b = read_excel(tf, 2)
#data_a$Coexistence[data_a$Coexistence==1] = 0.5
#data_a$`NPP (scaled)` = data_a$`NPP (scaled)`/2

ggplot(data_a, aes(y = data_a$Coexistence, x = data_a$`NPP (scaled)`)) +
  geom_point(alpha = 1/50, size = 3) +
  theme_bw() +
  geom_smooth(method='lm', linetype = "dashed", color = "black", size = 0.5) +
  xlab("NPP (re-scaled)") +
  ylab("Coexistence incidence") 

ggsave("figure_a.pdf")

# glm_tibble = tibble(Coexistance = data_b$X__1[2:5], slope = data_b$GLM[2:5], LCI = data_b$X__2[2:5], UCI = data_b$X__3[2:5], label = "GLM")
# pglmm_tibble = tibble(Coexistance = data_b$X__1[2:5], slope = data_b$PGLMM[2:5], LCI = data_b$X__4[2:5], UCI = data_b$X__5[2:5], label = "PGLMM")
# combined_tibble = bind_rows(glm_tibble, pglmm_tibble)
# combined_tibble
# ggplot(combined_tibble, aes(x = Coexistance, fill = label)) + 
#   geom_bar() +
#   xlab("Range overlap threshold (%)") + 
#   ylab("NP-Coexistance slope") + 
#   labs(color = "Cluster assignment")
# 
# ggsave("figure_b.pdf")
```
